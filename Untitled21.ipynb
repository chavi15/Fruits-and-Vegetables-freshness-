```python
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from PIL import Image

import tensorflow as tf
from sklearn.model_selection import train_test_split
from sklearn.utils.class_weight import compute_class_weight
from sklearn.metrics import classification_report, confusion_matrix

import kagglehub



```

```python
DATA_DIR = kagglehub.dataset_download("muhammad0subhan/fruit-and-vegetable-disease-healthy-vs-rotten")

print("Path to dataset files:", DATA_DIR)
image_paths = []
labels = []

dataset_folder_name = os.listdir(DATA_DIR)[0]
dataset_inner_path = os.path.join(DATA_DIR, dataset_folder_name)

for category_name in os.listdir(dataset_inner_path): # This should list 'Apple__Healthy', 'Grape__Rotten', etc.
    category_dir = os.path.join(dataset_inner_path, category_name)
    if os.path.isdir(category_dir):
        for img_file_name in os.listdir(category_dir):
            full_image_path = os.path.join(category_dir, img_file_name)
            if os.path.isfile(full_image_path): # Ensure we're adding actual files, not subdirectories
                image_paths.append(full_image_path)
                labels.append(category_name)

df = pd.DataFrame({
    "path": image_paths,
    "label": labels
})

df.head()
```

```python
print("Total images:", len(df))
df['label'].value_counts()
```

```python
df['label'].value_counts().plot(kind='bar', figsize=(10,4))
plt.title("Class Distribution")
plt.show()
```

```python
df['status'] = df['label'].apply(
    lambda x: 'Healthy' if 'Healthy' in x else 'Rotten'
)

df['status'].value_counts()
```

```python
df['status'].value_counts().plot(kind='bar', color=['green','red'])
plt.title("Healthy vs Rotten Distribution")
plt.show()
```

```python
sizes = []

for p in df['path'][:1000]:
    img = Image.open(p)
    sizes.append(img.size)

sizes = np.array(sizes)

print("Min size:", sizes.min(axis=0))
print("Max size:", sizes.max(axis=0))
```

```python
bad_images = []

for p in df['path']:
    try:
        Image.open(p).verify()
    except:
        bad_images.append(p)

print("Corrupted images:", len(bad_images))
```

```python
df = df[~df['path'].isin(bad_images)]
```

```python
train_df, val_df = train_test_split(
    df,
    test_size=0.2,
    stratify=df['status'],
    random_state=42
)
```

```python
IMG_SIZE = (224, 224)
BATCH_SIZE = 16

def load_image(path, label):
    img = tf.io.read_file(path)
    img = tf.image.decode_jpeg(img, channels=3)
    img = tf.image.resize(img, IMG_SIZE)
    img = img / 255.0
    return img, label
```

```python
train_ds = tf.data.Dataset.from_tensor_slices(
    (train_df['path'], train_df['status'].map({'Healthy':0, 'Rotten':1}))
)

val_ds = tf.data.Dataset.from_tensor_slices(
    (val_df['path'], val_df['status'].map({'Healthy':0, 'Rotten':1}))
)
```

```python
AUTOTUNE = tf.data.AUTOTUNE

train_ds = train_ds.map(load_image, num_parallel_calls=AUTOTUNE)
val_ds   = val_ds.map(load_image, num_parallel_calls=AUTOTUNE)

train_ds = train_ds.shuffle(1024).batch(BATCH_SIZE).prefetch(AUTOTUNE)
val_ds   = val_ds.batch(BATCH_SIZE).prefetch(AUTOTUNE)
```

```python
y_train = train_df['status'].map({'Healthy':0, 'Rotten':1}).values

class_weights = compute_class_weight(
    class_weight="balanced",
    classes=np.unique(y_train),
    y=y_train
)

class_weights = dict(enumerate(class_weights))
class_weights
```

```python
data_augmentation = tf.keras.Sequential([
    tf.keras.layers.RandomFlip("horizontal"),
    tf.keras.layers.RandomRotation(0.15),
    tf.keras.layers.RandomContrast(0.2)
])
```

```python
base_model = tf.keras.applications.EfficientNetB2(
    input_shape=(224,224,3),
    include_top=False,
    weights="imagenet"
)

base_model.trainable = False
```

```python
inputs = tf.keras.Input(shape=(224,224,3))
x = data_augmentation(inputs)
x = base_model(x, training=False)
x = tf.keras.layers.GlobalAveragePooling2D()(x)
x = tf.keras.layers.BatchNormalization()(x)
x = tf.keras.layers.Dense(256, activation='relu')(x)
x = tf.keras.layers.Dropout(0.4)(x)
outputs = tf.keras.layers.Dense(1, activation='sigmoid')(x)

model = tf.keras.Model(inputs, outputs)
```

```python
model.compile(
    optimizer=tf.keras.optimizers.Adam(1e-3),
    loss='binary_crossentropy',
    metrics=['accuracy']
)
```

```python
callbacks = [
    tf.keras.callbacks.EarlyStopping(
        monitor='val_loss',
        patience=3,
        restore_best_weights=True
    )
]
```

```python
history = model.fit(
    train_ds,
    validation_data=val_ds,
    epochs=20,
    callbacks=callbacks,
    class_weight=class_weights
)
```

```python
base_model.trainable = True

for layer in base_model.layers[:-30]:
    layer.trainable = False
```

```python
model.compile(
    optimizer=tf.keras.optimizers.Adam(1e-5),
    loss='binary_crossentropy',
    metrics=['accuracy']
)
```

```python
history_ft = model.fit(
    train_ds,
    validation_data=val_ds,
    epochs=10,
    callbacks=callbacks,
    class_weight=class_weights
)
```

```python
y_true = np.concatenate([y.numpy() for _, y in val_ds])
y_pred = (model.predict(val_ds) > 0.5).astype(int)

print(classification_report(y_true, y_pred))
```
